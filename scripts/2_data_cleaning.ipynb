```python
import pandas as pd
import glob
import re
path="./../data/raw/*"
def file_list():
    files = glob.glob(path)
    new_files = []

    for file in files:
        file=file.replace("\\","/")
        print(f"Found file: {file}")
        new_files.append(file)
    return new_files

x=file_list()
print(x)

for file in x:

    print(f"Processing file: {file}")
    data=pd.read_csv(file,encoding='latin1',index_col=False)
    print("Columns: ", data.columns if hasattr(data, 'columns') else "No columns found")

    print(data.dtypes)
    if 'InvoiceDate' in data.columns:
        data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], errors='coerce')
    if 'CustomerID' in data.columns:
        data['CustomerID'] = pd.to_numeric(data['CustomerID'], errors='coerce')
    else:
        print("Warning: 'InvoiceDate' column not found in this dataset.")


    data.columns=[col.replace("ï»¿","") for col in data.columns]
    data.columns=[re.sub(r'(?<!^)(?=[A-Z])','_', col).lower() for col in data.columns]
    data.columns=[col.replace("_i_d","_id") for col in data.columns]

    print(f"Processing file: {file}")
    print("Columns: ", data.columns if hasattr(data, 'columns') else "No columns found")
    data.to_csv(file.replace("raw","cleaned").replace(".csv","_cleaned.csv"),index=False)
    print(f"File {file} has {data.shape[0]} rows and {data.shape[1]} columns.")
    print(data.describe(include='all'))

```

```python
print("let's start")
import pandas as pd
import glob


path="./../data/raw/*"
def file_list():
    files = glob.glob(path)
    new_files = []

    for file in files:
        file=file.replace("\\","/")
        print(f"Found file: {file}")
        new_files.append(file)
    return new_files

x=file_list()
print(x)

def file_overview():
    print("Overview of files:")
    for file in x:
        print(f"Processing file: {file}")
        data=pd.read_csv(file,encoding='latin1')
        print(f"File {file} has {data.shape[0]} rows and {data.shape[1]} columns.")
        print(data.describe(include='all')) 
        print("\n Lack of values: ")
        print(data.isnull().sum())

file_overview()
```
